```python
a = "123456789abcd"
print(len(a))
n = 50
step = len(a) // n
if step == 0:
    step = 1
    n = len(a)
print(step)
b = []
for i in range(n-1):
    a1 = a[i*step:(i+1)*step]
    a2 = a[0:i*step]+a[(i+1)*step:]
    b.append((a1,a2))
a1 = a[(n-1)*step:]
a2 = a[0:(n-1)*step]
b.append((a1,a2))
print(b)
```
